```python
import cv2
from PIL import Image
```

```python
img = cv2.imread("clear_image.jpg",flags=cv2.IMREAD_GRAYSCALE)
```

```python
img
```

```python
_, new_img = cv2.threshold(img,150,255,cv2.THRESH_BINARY)
```

```python
new_img
```

```python
Image.fromarray(new_img)
```

### Trying simple and adaptive tresholding for darker image

```python
dar_img = cv2.imread("dark_image.jpg",flags=cv2.IMREAD_GRAYSCALE)
dar_img
```

```python
_, i = cv2.threshold(dar_img,150,255,cv2.THRESH_BINARY)
i
```

```python
Image.fromarray(i)
```

## Use adaptive threshold

```python
new_img = cv2.adaptiveThreshold(
    dar_img,255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    61,
    11
)
```

```python
Image.fromarray(new_img).show()
```
